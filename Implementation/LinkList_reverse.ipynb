```python
"""Reverse a singly linked list.

Example:

Input: 1->2->3->4->5->NULL
Output: 5->4->3->2->1->NULL
Follow up:

A linked list can be reversed either iteratively or recursively. Could you implement both?"""

# Algorithm
# create 2 pointer prev and curr (prev holds previous node, curr hold curent node)
# prev = None , curr = head
# while curr ! = None

# next_node = curr.next
# curr.next = prev, prev = curr, curr = next_node
# return prev
```

```python
#Create a node with data and next value pointer
class Node:
    def __init__(self, data = None):
        self.data = data
        self.next_node = None
```

```python
#Holds functions for creating, displaying and inserting into a link list 
class LinkList:
    #Constructor for linklist, initialse head of linklist to nothing
    def __init__(self):
        self.head = None
    
    def display_ll(self):
        if(self.head == None):
            print("Nothing to print")
        else:
            temp = self.head
            while(temp!=None):
                print(temp.data)
                temp = temp.next_node
                
    
    
    def reverseList(self, head: ListNode) -> ListNode:
        prev, curr = None, head
        
        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        
        return prev
    
    
    
    
    def insert_linklist(self,data):
        newdata = Node(data)
        if(self.head == None):
            self.head = newdata

        else:
            temp = self.head
            while(temp.next_node != None):
                temp = temp.next_node
            temp.next_node = newdata
            
    def reverse_linklist(self):
        #If no elements are present in the link list
        if (self.head == None):
            print("Nothing to reverse")
        current = self.head
        #If there is just one element in the link list 
        if(current.next_node == None):
            print(current)
        else:
            next_current = current.next_node
        #If only 2 elements are present in the link list
            if(next_current.next_node == None):
                current.next_node = None
                next_current.next_node = current
                self.head = next_current
            else:
                next_next_current = next_current.next_node
                next_current.next_node = current
                current.next_node = None
                while(next_next_current.next_node):
                    next_current.next_node = current
                    current = next_current
                    next_current = next_next_current
                    next_next_current = next_current.next_node
                    next_current.next_node = current
                next_next_current.next_node = next_current
                self.head = next_next_current
                
    def reverse_linklist_pairs(self, start, pairs):
        if(self.head == None):
            print("Nothing to reverse")
        current = self.head
        #If there is just one element in the link list 
        if(current.next_node == None):
            print(current)
        else:
            next_current = current.next_node
        #If only 2 elements are present in the link list
            if(next_current.next_node == None and pairs < 3):
                current.next_node = None
                next_current.next_node = current
                self.head = next_current
            else:
                next_next_current = next_current.next_node
                count = 1
                while(next_next_current.next_node!= None and count < pairs ):
                    next_current.next_node = current
                    current = next_current
                    next_current = next_next_current
                    next_next_current = next_current.next_node
                    next_current.next_node = current
                    count += 1
                    print("Current " ,current.data)
                    print("Next " ,next_current.data)
                    print("Next Next " ,next_next_current.data)
                next_next_current.next_node = next_current
                self.head = next_next_current 
                
                
                
    def reverse(self, head, k): 
        current = head  
        next  = None
        prev = None
        count = 0 
          
        # Reverse first k nodes of the linked list 
        while(current is not None and count < k): 
            next = current.next
            current.next = prev 
            prev = current 
            current = next 
            count += 1
  
        # next is now a pointer to (k+1)th node 
        # recursively call for the list starting 
        # from current. And make rest of the list as 
        # next of first node 
        if next is not None: 
            head.next = self.reverse(next, k) 
  
        # prev is new head of the input list 
        return prev 
```

```python
# Definition for singly-linked list.
class ListNode:
     def __init__(self, x):
         self.val = x
         self.next = None
```

```python
class Solution:
    def reverseList(head):
        if head == None:
            return None
        curr = head
        if(curr.next):
            curr_next = curr.next
            head.next = None
        else:
            return head
        if(curr_next.next):
            curr_next_next = curr_next.next
            while(curr_next_next):
                temp = curr_next_next
                curr_next.next = curr
                curr = curr_next
                curr_next = temp
                curr_next_next = curr_next.next
            curr_next.next = curr
            head = curr_next
        else:
            head = curr_next
            curr_next.next = curr
        return head
```
