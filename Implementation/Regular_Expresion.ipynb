```python
import itertools
def solution(exp):
    integer_list = [num for num in exp if num.isdigit()]
    
    output = []
    dict_num = dict()
    exp_count = 0
    for i in range(len(exp)):
        dict_num[i] = exp[i]
        if exp[i] == '$':
            exp_count += 1
    
    seq = ''.join(integer_list)
    prod = [list(x) for x in itertools.product(seq, repeat = exp_count)]
    print(prod)
    
    
    for p in prod:
        nu = ""
        for key, val in dict_num.items():
            if val != '$':
                nu += val
            else:
                nu += p.pop(0)
        output.append(nu.lstrip('0'))
                
    print(output)  

    
exp = '23$$3'
print(solution(exp))
```
