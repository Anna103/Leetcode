```python
# Print pre-order, inorder, post order traversal of biary tree

# pre-order = Root Left Right
# inorder   = Left Root Right
# postorder = Left Right Root
```

```python
class Tree:    
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
    
    def insert(self, new_node):
        #No element in the Node
        if self.data == None:
            self.data = new_node
        else:
            if self.data > new_node.data:
                if self.left == None:
                    self.left = new_node
                else:
                    self.left.insert(new_node)
            else:
                if self.right == None:
                    self.right = new_node
                else:
                    self.right.insert(new_node)

    def inorder(self): 
        if self.left: 
            self.left.inorder() 
        print(self.data) 
        if self.right:
            self.right.inorder() 
    
    def inorder_iteration(self):
        data_stack = []
        data_stack_2 = []
        data_stack.append(self)
        s_left = self.left
        s_right = self.right
        while(s_left):
            data_stack.append(s_left)
            s_left = s_left.left
        while(data_stack):
            current = data_stack.pop()
            data_stack_2.append(current.data)
            print("Data Stack")
            for i in range(len(data_stack)):
                    print(data_stack[i].data)
            print("Data Stack 2",data_stack_2)
            if(current.right != None):
                data_stack.append(current.right)
                s_left = current.left
                if(s_left.data not in data_stack_2):
                    while(s_left):
                        data_stack.append(s_left)
                        s_left = s_left.left
                #for i in range(len(data_stack)):
                #    print(data_stack[i].data)
                    
        print(data_stack_2)
    def preorder(self): 
        if self: 
            print(self.data)
        if self.left:    
            self.left.preorder()  
        if self.right:
            self.right.preorder() 
            
    def postorder(self): 
        if self.left:    
            self.left.postorder()  
        if self.right:
            self.right.postorder()
        if self: 
            print(self.data)
```

```python

r = Tree(50)
r.insert(Tree(30)) 
r.insert(Tree(20)) 
r.insert(Tree(40)) 
r.insert(Tree(70)) 
r.insert(Tree(60)) 
r.insert(Tree(80)) 
print("Inorder")
r.inorder() 


print("Inorder Iteration")
r.inorder_iteration()

print("Pre Order")
r.preorder()

print("Post Order")
r.postorder()
```
