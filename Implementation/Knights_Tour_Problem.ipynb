```python
class Solution:
    
    #global step_number
    
    # This function will check if it is safe to make a move in that cell
    def isSafe(self, i, j, chessboard):
        return((i >= 0 and i < 4) and (j >= 0 and j < 4) and chessboard[i][j] ==-1)
    
    def backtracking(self, board, i, j, step_count):
        
        # The below lists contain all the valid moves allowed by the knight
        valid_x = [2, 1, -1, -2, -2, -1, 1, 2] 
        valid_y = [1, 2, 2, 1, -1, -2, -2, -1] 
        
        if (step_count == 4**2):
            return True
        for k in range(len(valid_x)):
            if(self.isSafe(i+valid_x[k], j+valid_y[k], board)):
                board[i+valid_x[k]][j+valid_y[k]] = step_count
                print(board)
                if(self.backtracking(board, i+valid_x[k], j+valid_y[k], step_count+1)):
                    return True
                board[i+valid_x[k]][j+valid_y[k]]= -1
        return False
                
                    
                
    # This function uses the backtracking function and checks if the solution is available to given problem
    # It creates the boards initial status with all the cells as -1
    def knightPro(self):
        
        board = [[-1 for cols in range(4)]for rows in range(4)]
        
        # First position is numbered as 0, since we being from this place
        board[0][0] = 0
        
        # We have made our first move hence the step_count will become 1
        step_count = 1
        
        if(not self.backtracking(board, 0, 0, step_count)):
            print("Solution doesnot exist")
        else:
            print(board)
        
```
