```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinklistReverser:
    
    def reverse_linklist(self, node):
        prev = None
        curr = node
        old_head = node
        node_dict = dict()
        while curr:
            node_dict[curr] = curr.data
            next_curr = curr.next
            if(next_curr not in node_dict.keys()):
                curr.next = prev
                prev = curr
                curr = next_curr
            else:
                curr.next = prev
                old_head.next = curr
                return curr
        
    
    def print_linklist(self, node):
        temp = node
        while(temp):
            print(temp.data)
            temp = temp.next
```

```python
if __name__ == "__main__":
    node1 = Node(1)
    node2 = Node(2)
    node1.next = node2
    node3 = Node(3)
    node2.next = node3
    node4 = Node(4)
    node3.next = node4
    node5 = Node(5)
    node4.next = node5
    node5.next = node1
    
    llr = LinklistReverser()
    rev = llr.reverse_linklist(node1)
    llr.print_linklist(rev)
```
