```python
# Concept of binary search is used.
# reference : https://www.youtube.com/watch?v=LPFhl65R7ww


# array 1 is always smaller than array 2
# Our aim here would be to find a partition such that 
# X1 | X2 ====== array 1
# Y1 | Y2 ====== array 2

# All elements on left should be lesser than all elemenets on right
# Start = 0
# End = length of Array 1

# X = len(Array1)
# Y = len(Array2)

# PartitionX = (start + end) // 2
# PartitionY = ((X+ Y + 1 )//2) - PartitionX


#If above any of above partition is 0 then X1 or Y1 will be -inf (Becoz nothing on the left)

#else
    # X1 = element at PartitionX - 1 position
    # Y1 = element at PartitionY - 1 position

#If above any of above partition is len of that array  then X2 or Y2 will be inf (Becoz nothing on the right) 

#else
    # X2 = element at PartitionX position
    # Y2 = element at PartitionY position

    
# If elements on right is bigger, then check is sum of lengths of array is even
#  If yes then media is (max(X1, Y1) + min (X2, Y2) ) / 2
#    else median is max(X1, Y1)




# Else if Y1 > X2
# start = partitionX + 1 i.e move partion to right of X

#else
# start = partitionX - 1 i.e move partion to left of X
```

```python
class Solution:
    
    def median_answer(self, array_1, array_2):
        
        # Find the shortest of 2 list in the given 2 array,
        # Binary search is performed in smallest array
        
        if(len(array_1) > len(array_2)):
            self.median_answer(array_2, array_1)
        
        start = 0
        end = len(array_1)
        X = len(array_1)
        Y = len(array_2)
        while(start <= end):
            
            partitionX = int((start + end )/2)
            partitionY = int((X + Y + 1 )/2 - partitionX)
            
            
            # Edge case when there is nothing on the left side, then we assign x1 to infinity
            if(partitionX == 0):
                X1 = float('-inf')
            else:
                X1 = array_1[partitionX - 1]
                
            if(partitionX == len(array_1) ):
                X2 = float('inf')
            else:
                X2 = array_1[partitionX]
                
            if(partitionY == 0):
                Y1 = float('-inf')
            else:
                Y1 = array_2[partitionY - 1]
                
            if(partitionY == len(array_2) ):
                Y2 = float('inf')
            else:
                Y2 = array_2[partitionY]
            
            print(X1, X2, Y1, Y2)
            if((X1 <= Y2) and (Y1 <= X2)):
                # We have found correct partitions
                
                #Check if the sum of length of both is odd or even
                
                if( (X+Y) % 2 == 0):
                    print((max(X1,Y1) + min(X2, Y2))/2)
                else:
                    print(max(X1,Y1))
                break
            elif(Y1 > X2):
                start = partitionX + 1
            else:
                end = partitionX - 1
```
