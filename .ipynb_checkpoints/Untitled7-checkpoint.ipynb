```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
readcsv=pd.read_csv("/Users/admin/Desktop/Fall2018/Machine Learning/Kaggle/StudentsPerformance.csv")
```

```python
readcsv.describe()
```

```python
readcsv.head(10)
```

```python
grade = pd.DataFrame(np.nan,index=range(0,len(readcsv)) ,columns=['A'])
for i in range(readcsv.shape[0]):
    if(readcsv['math score'][i] <40 or readcsv['reading score'][i]<40 or readcsv['writing score'][i]<40):
        grade[i]='F'
    else:
        if((readcsv['math score'][i]+readcsv['reading score'][i]+readcsv['writing score'][i])/3>=80):
            grade[i]='A'
        elif((readcsv['math score'][i]+readcsv['reading score'][i]+readcsv['writing score'][i])/3>=60):
            grade[i]='B'
        else:
            grade[i]='C'
grade=grade.drop(columns=['A'])
#print(grade[:1])
readcsv['Grade']=grade[:1].T
print(readcsv)           
```

```python
readcsv.head(10)
```

```python
sns.set(style="darkgrid")
p = sns.countplot(x='parental level of education', data = readcsv, hue='Grade', palette='bright')
axn = plt.setp(p.get_xticklabels(), rotation=90)
```

```python
sns.set(style="darkgrid")
p = sns.countplot(x='race/ethnicity', data = readcsv, hue='Grade', palette='bright')
axn = plt.setp(p.get_xticklabels(), rotation=90)
```
