```python
class TicTacToe:
    
    global board 
    board = [['-','-','-'],['-','-','-'],['-','-','-']]
    #This method will add the x or 0 to the board
    def add_token(self,row, column, value):
        if(board[row][column] == "-"):
            board[row][column] = value
            if(self.check_win()):
                print("Game Won")
                self.print_board()
                raise Exception("Won the game")
            self.print_board()
        else:
            print("The location is invalid")
    
    #This method will print the board
    def print_board(self):
        print("\n")
        for i in range(len(board)):
            print(board[i][0] + "|" + board[i][1]+ "|" + board[i][2])
        if(self.check_win()):
                print("Game Won")
                raise Exception("Won the game")
    
    #This method will let AI to play its part.
    #If board is full it will raise and exception and end the game
    def board_move(self):
        for i in range(len(board)):
            for j in range(len(board[i])):
                if(self.check_win()):
                    self.print_board()
                    raise Exception("Game Won")
                if(board[i][j] == "-"):
                    board[i][j] = "O"
                    self.print_board()
                    return board
        
        raise Exception("Board Full")
    
    # Every time a move is played
    #Check if board is full
    
    def check_win(self):
        for i in range(len(board)):
            if(board[i][0] == board[i][1] and board[i][1] == board[i][2] and board[i][0] != '-'):
                print("1",board[i][0],board[i][1],board[i][2])
                return True
        for i in range(len(board)):
            if(board[0][i] == board[1][i] and board[0][i] == board[2][i] and board[0][i] != '-'):
                print("2")
                return True    
        if(board[0][0] == board[1][1] and board[1][1] == board[2][2] and board[0][0] != '-'):
            print("3")
            return True
        if(board[0][2] == board[1][1] and board[0][2] == board[2][2] and board[0][2] != '-'):
            print("4")
            return True
        return False
            
        

                    
                
                
```
